```
from pyspark.ml.feature import VectorAssembler

datasetDF = sqlContext.sql("SELECT * FROM plant_dataset WHERE record='EFConsPerCap'")

vectorizer = VectorAssembler()
#vectorizer.setInputCols(["year", "crop_land", "grazing_land", "forest_land", "fishing_ground", "built_up_land", "total", "Percapita GDP (2010 USD)", "population"])
vectorizer.setInputCols(["year", "total", "Percapita GDP (2010 USD)", "population"])
vectorizer.setOutputCol("features")

datasetDF.count()
```

```
#(split15DF, split85DF) = datasetDF.randomSplit([.15, .85], 1900009193L)
#testSetDF = split15DF.cache()
#trainingSetDF = split85DF.cache()
datasetDF_notNull = datasetDF.filter(datasetDF['carbon'].isNotNull())
#datasetDF_notNull.show()
datasetDF_notNull = datasetDF_notNull.filter(datasetDF_notNull["Percapita GDP (2010 USD)"].isNotNull())
datasetDF_notNull = datasetDF_notNull.filter(datasetDF_notNull["population"].isNotNull())
datasetDF_notNull = datasetDF_notNull.filter(datasetDF_notNull["total"].isNotNull())

testSetDF = datasetDF_notNull.filter(datasetDF_notNull['year']<2011)
trainingSetDF = datasetDF_notNull.filter(datasetDF_notNull['year']>=2011)
print(testSetDF.count())
print(trainingSetDF.count())
```

```
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.tuning import ParamGridBuilder, CrossValidator
from pyspark.ml import Pipeline
from pyspark.ml.regression import RandomForestRegressor

# Create a RandomForestRegressor
rf = RandomForestRegressor()

rf.setPredictionCol("Prediction_carbon").setFeaturesCol("features").setSeed(190088121L).setMaxDepth(8).setNumTrees(25).setLabelCol("carbon")
#setPredictionCol("Prediction_PE").setFeaturesCol("features").setLabelCol("PE").setMaxBins(100)

# Create a Pipeline
rfPipeline = Pipeline()

# Set the stages of the Pipeline
rfPipeline.setStages([vectorizer, rf])
```

```
# TODO: Replace <FILL_IN> with the appropriate code.
# Let's just reuse our CrossValidator with the new rfPipeline,  RegressionEvaluator regEval, and 3 fold cross validation
regEval = RegressionEvaluator(predictionCol="Prediction_carbon", labelCol="carbon", metricName="rmse")
crossval = crossval = CrossValidator(estimator=rfPipeline, evaluator=regEval, numFolds=3)

# Let's tune over our rf.maxBins parameter on the values 50 and 100, create a paramter grid using the ParamGridBuilder
paramGrid = ParamGridBuilder().addGrid(rf.maxBins, [50, 100]).build()

# Add the grid to the CrossValidator
crossval.setEstimatorParamMaps(paramGrid)

# Now let's find and return the best model
rfModel = crossval.fit(trainingSetDF).bestModel
```

```
# TODO: Replace <FILL_IN> with the appropriate code.

# Now let's use rfModel to compute an evaluation metric for our test dataset: testSetDF
resultsDF = rfModel.transform(testSetDF)

# Run the previously created RMSE evaluator, regEval, on the resultsDF DataFrame
rmseRF = regEval.evaluate(resultsDF)

# Now let's compute the r2 evaluation metric for our test dataset
r2RF = regEval.evaluate(resultsDF, {regEval.metricName: 'r2'})

print("RF Root Mean Squared Error: {0:.2f}".format(rmseRF))
print("RF r2: {0:.2f}".format(r2RF))
```
