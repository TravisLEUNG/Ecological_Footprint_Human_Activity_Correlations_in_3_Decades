```
from pyspark.ml.feature import VectorAssembler

datasetDF = sqlContext.sql("SELECT * from carbon_dataset where record = 'EFConsTotGHA'")

vectorizer = VectorAssembler()
vectorizer.setInputCols(["year", "total", "Percapita GDP (2010 USD)", "population"])
vectorizer.setOutputCol("features")
```

```
datasetCarbonNotNullDF = datasetDF.filter(datasetDF['carbon'].isNotNull()).filter(datasetDF['population'].isNotNull()).filter(datasetDF['Percapita GDP (2010 USD)'].isNotNull()).cache()

trainingSetDF = datasetCarbonNotNullDF.filter(datasetCarbonNotNullDF['year']<2011).cache()
testSetDF = datasetCarbonNotNullDF.filter(datasetCarbonNotNullDF['year']>=2011).cache()

trainingSetDF.count()+testSetDF.count()
```

```
# ***** LINEAR REGRESSION MODEL ****

from pyspark.ml.regression import LinearRegression
from pyspark.ml.regression import LinearRegressionModel
from pyspark.ml import Pipeline

# Let's initialize our linear regression learner
lr = LinearRegression()

# We use explain params to dump the parameters we can use
print(lr.explainParams())
```

```
# Now we set the parameters for the method
lr.setPredictionCol("Prediction_Carbon")\
  .setLabelCol("carbon")\
  .setMaxIter(100)\
  .setRegParam(0.15)


# We will use the new spark.ml pipeline API. If you have worked with scikit-learn this will be very familiar.
lrPipeline = Pipeline()

lrPipeline.setStages([vectorizer, lr])

# Let's first train on the entire dataset to see what we get
lrModel = lrPipeline.fit(trainingSetDF)
```

```
# The intercept is as follows:
intercept = lrModel.stages[1].intercept

# The coefficents (i.e., weights) are as follows:
weights = lrModel.stages[1].coefficients

# Create a list of the column names
templist = ["year", "total", "Percapita GDP (2010 USD)", "population"]
featuresNoLabel = [col for col in datasetDF.columns if col in templist]

# Merge the weights and labels
coefficents = zip(weights, featuresNoLabel)

# Now let's sort the coefficients from greatest absolute weight most to the least absolute weight
coefficents.sort(key=lambda tup: abs(tup[0]), reverse=True)

equation = "y = {intercept}".format(intercept=intercept)
variables = []
for x in coefficents:
    weight = abs(x[0])
    name = x[1]
    symbol = "+" if (x[0] > 0) else "-"
    equation += (" {} ({} * {})".format(symbol, weight, name))

# Finally here is our equation
print("Linear Regression Equation: " + equation)
```

```
# Apply our LR model to the test data and predict power output
resultsDF = lrModel.transform(testSetDF).select("population", "carbon", "Prediction_Carbon")

display(resultsDF)
```

```
# Now let's compute an evaluation metric for our test dataset
from pyspark.ml.evaluation import RegressionEvaluator

# Create an RMSE evaluator using the label and predicted columns
regEval = RegressionEvaluator(predictionCol="Prediction_Carbon", labelCol="carbon", metricName="rmse")

# Run the evaluator on the DataFrame
rmse = regEval.evaluate(resultsDF)

print("Root Mean Squared Error: %.2f" % rmse)
```

```
# Now let's compute another evaluation metric for our test dataset
r2 = regEval.evaluate(resultsDF, {regEval.metricName: "r2"})

print("r2: {0:.2f}".format(r2))
```

```
# First we remove the table if it already exists
sqlContext.sql("DROP TABLE IF EXISTS Carbon_RMSE_Evaluation")
dbutils.fs.rm("dbfs:/user/hive/warehouse/Carbon_RMSE_Evaluation", True)

# Next we calculate the residual error and divide it by the RMSE
resultsDF.selectExpr("carbon", "Prediction_Carbon", "carbon - Prediction_Carbon Residual_Error", "(carbon - Prediction_Carbon) / {} Within_RSME".format(rmse)).registerTempTable("Carbon_RMSE_Evaluation")
```
